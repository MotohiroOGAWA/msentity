# Basic Usage of msentity

This notebook demonstrates how to use `msentity` for loading, inspecting,
filtering, and writing mass spectrometry datasets.

## Loading a dataset

Datasets can be loaded from MSP or MGF files.

```python
from msentity.io.msp import read_msp

dataset = read_msp("example.msp")
print(dataset)
```

## Inspecting metadata

You can access dataset-level metadata in a pandas-like way.

```python
print(dataset["Name"])
print(dataset["PrecursorMZ"])
```

## Iterating over spectra

You can iterate over records in the dataset.

```python
for record in dataset:
    print(record["name"], record.n_peaks)
```

## Accessing a single record

```python
record = dataset[0]

print(record)
print(record["name"])
print(record.n_peaks)
```

## Accessing spectral peaks

```python
spectrum = record.spectrum

print(spectrum.mz)
print(spectrum.intensity)
```

## Filtering the dataset

You can filter using boolean masks.

```python
subset = dataset[dataset["precursor_mz"] > 300]
print(subset["name"])
```

## Sorting

```python
sorted_dataset = dataset.sort_by("precursor_mz", ascending=False)
print(sorted_dataset["precursor_mz"])
```

## Writing datasets

Datasets can be written back to MSP or MGF files.

```python
from msentity.io.msp import write_msp
from msentity.io.mgf import write_mgf

write_msp(dataset, "output.msp")
write_mgf(dataset, "output.mgf")
```

## Typical workflow

A common data analysis pattern.

```python
dataset = read_msp("example.msp")

filtered = dataset[dataset["precursor_mz"] > 300]
filtered = filtered.sort_by("precursor_mz", ascending=False)

for record in filtered:
    print(record["name"], record["precursor_mz"], record.n_peaks)
```
